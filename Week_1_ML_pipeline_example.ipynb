```python
# Importing necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# Load the Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)

display(X)

y = iris.target

# Check the first few rows of the dataset
print(X.head())

# Preprocessing pipeline with ColumnTransformer
# We will scale the numerical columns (sepal length, sepal width, petal length, petal width)
# Since there are no categorical columns here, the 'remainder' will pass unchanged.

# Step 1: Define the column transformer
preprocessor = ColumnTransformer(
    transformers=[('num', StandardScaler(), X.columns)], 
    remainder='passthrough'  # Any other columns that are not listed will be passed through unchanged
)
display(preprocessor)

# Step 2: Create a pipeline with the preprocessor and a classifier (RandomForest)
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())  # Classifier step
])

# Step 3: Fit the pipeline to the data
pipeline.fit(X, y)

# Step 4: Make predictions using the trained model
predictions = pipeline.predict(X)
print(predictions)  # Show first 5 predictions
```
